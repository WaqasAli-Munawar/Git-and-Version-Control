In this file, we'll walk through the process of setting Git up on our own machine and authenticating with GitHub. Afterwards, we'll be able to sync the changes we make to data science projects locally with GitHub. Finally, when we're ready, we can publish our code to GitHub and build a portfolio of projects for others to see.

Fortunately, installing Git only involves a few steps.

1. Navigate to the [Git downloads page](https://git-scm.com/downloads) and download the appropriate installer for our operating system.
2. Run the installer and step through the installation wizard.
3. Open the command line program (Terminal for Linux and Mac and Command Prompt for Windows) and run git version.

If the output describes the current Git version and doesn't throw an error, we've successfully installed Git!

Now let's configure Git with our name and email address. Run the following shell command, but replace "OUR NAME" with our actual name:

* git config --global user.name "OUR NAME"

Then, run the following shell command, but replace "OUR EMAIL ADDRESS" with our actual email address:

* git config --global user.email "OUR EMAIL ADDRESS"

Use a Web browser to navigate to [GitHub](https://github.com/) and create an account. There are three main steps we'll have to complete:

* Create a personal account. Select a unique username and password and enter our email.
* Choose a plan. If we select the free plan, all of our code (which is organized in repositories) will be public. Select the free plan for now. We can always upgrade to a paid plan later on, which would allow us to have private repositories.
* Read the GitHub [Hello World guide](https://guides.github.com/activities/hello-world/)
.
Complete the instructions in step 1 of the Hello World guide to create our first repository on GitHub.

Now we need to authenticate our computer with GitHub so we can push code to our remote repositories. The easiest way to do this is to clone the repo we created on GitHub to our local computer. Click the **Clone or Download** button on the repository page, and then click **Use HTTPS** in the floating window to reveal the clone URL:



Then, run git clone {url} from the command line (replace {url} with the clone URL). Upon running this command, Git will download repository to our local computer as a new folder (if there are any files in the repo). The local repo will be "read-only" in the sense that we won't be able to push our changes back to the remote repo. We will still able to modify files and commit changes locally.

To verify that we can now push changes to our GitHub repo:

* Create a branch (git checkout -b testbranch).
* Make a small change (use **nano** if we want to edit a file from the command line).
* Commit the change (git commit -am 'test').
* Push the branch to GitHub (git push origin testbranch).
* By default, we'll be asked to log in every time our run git push, git pull, or another Git command to the remote repository. We can use a credential helper to get around this, however. To learn more about this option, see the [GitHub documentation](https://help.github.com/en/github/using-git/which-remote-url-should-i-use#cloning-with-https-urls-recommended).
